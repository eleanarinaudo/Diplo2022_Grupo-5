```python
df_to_g = filtered_df_by_exp_age_bruto.rename(
    columns={
        "profile_years_experience": "Años de Experiencia",
        "salary_monthly_BRUTO": "Media de Salario Mensual",
        "profile_studies_level": "Nivel de Estudios",
        "work_province": "Ubicacion",
    }
)

fig, axes = plt.subplots(ncols=2, figsize=(30, 12))
sns.set_theme(style="whitegrid", palette="icefire")
sns.barplot(
    ax=axes[0],
    data=df_to_g,
    x="Nivel de Estudios",
    y="Media de Salario Mensual",
    hue="Ubicacion",
).set_title("Salario por nivel de estudio", fontsize=25)
sns.set(font_scale=4)
sns.set_theme(style="whitegrid", palette="icefire")
sns.barplot(
    ax=axes[1],
    data=df_to_g,
    x="Años de Experiencia",
    y="Media de Salario Mensual",
    hue="Ubicacion",
).set_title(
    "Salarios para Buenos Aires vs Resto del Pais por Años de experiencia", fontsize=20
)
plt.ylabel("")

axes[0].legend()
sns.despine()
```

```python
univ_sal = df_to_g.loc[df_to_g["Nivel de Estudios"].isin(["Universitarios"])]
univ_sal = univ_sal.loc[univ_sal["Ubicacion"].isin(["Buenos Aires"])]
univ_sal = univ_sal["Media de Salario Mensual"].mean()
univ_sal = round(univ_sal, 2)

ter_sal = df_to_g.loc[df_to_g["Nivel de Estudios"].isin(["No Universitarios"])]
ter_sal = ter_sal.loc[ter_sal["Ubicacion"].isin(["Buenos Aires"])]
ter_sal = ter_sal["Media de Salario Mensual"].mean()
ter_sal = round(ter_sal, 2)

print("Universitarios:", univ_sal, "No Universitarios:", ter_sal)

avrg_difff = (univ_sal - ter_sal) * 100 / univ_sal
avrg_difff = round(avrg_difff, 2)
print(
    "Los estudiantes Universitarios ganan un ",
    avrg_difff,
    "% más que los No Universitarios en Buenos Aires",
)
```

```python
ter_salb = df_to_g.loc[df_to_g["Nivel de Estudios"].isin(["No Universitarios"])]
ter_salb = ter_salb.loc[ter_salb["Ubicacion"].isin(["Buenos Aires"])]
ter_salb = ter_salb["Media de Salario Mensual"].mean()
ter_salb = round(ter_salb, 2)

univ_ = df_to_g.loc[df_to_g["Nivel de Estudios"].isin(["Universitarios"])]
univ_ = univ_.loc[univ_["Ubicacion"].isin(["Resto del país"])]
univ_ = univ_["Media de Salario Mensual"].mean()
univ_ = round(univ_, 2)

print(
    "Salario para No Universitario en Buenos Aires:",
    ter_salb,
    "\nSalario para Universitarios en el Resto de País: ",
    univ_,
)
```

```python
a = df_to_g.loc[df_to_g["Años de Experiencia"].isin(["Mayor o igual a 5"])]
a = a.loc[a["Ubicacion"].isin(["Buenos Aires"])]
a = a["Media de Salario Mensual"].mean()
a = round(a, 2)

b = df_to_g.loc[df_to_g["Años de Experiencia"].isin(["Mayor o igual a 5"])]
b = b.loc[b["Ubicacion"].isin(["Resto del país"])]
b = b["Media de Salario Mensual"].mean()
b = round(b, 2)

print("Más de 5 Buenos Aires:", a, "\nMás de 5 Resto del país:", b)

avrg_difff = (univ_sal - ter_sal) * 100 / univ_sal
avrg_difff = round(avrg_difff, 2)
print(
    "Los estudiantes Universitarios con más de 5 años de experiencia ganan un",
    avrg_difff,
    "% mas que sus pares en el resto del pais",
)
```

A continuación, intentaremos descubrir si las variables son independientes determinando si la probabilidad de ganar más que la media aumenta siendo por un lado de la subpoblación de universitarios y por el otra de terciarios.


**Probabilidad condicional `salary_monthly_NETO` por Universitarios**

```python
avr_NETO = filtered_df.salary_monthly_NETO.mean()
prob_above_avg = len(
    filtered_df[filtered_df['salary_monthly_NETO'] >= avr_NETO]) / len(filtered_df)

is_above_avg_UNIV = filtered_df['salary_monthly_NETO'] > avr_NETO
univ_bool = filtered_df.profile_studies_level == 'Universitario'

univ_num = filtered_df.loc[filtered_df.profile_studies_level.isin([
                                                                    'Universitario'])]
univ_num = univ_num.profile_studies_level.count()

p_condic_above_avg = len(filtered_df[is_above_avg_UNIV & univ_bool])/univ_num
print(
    "Probabilidad Condicional Salario Neto por encima de la Media [Universitario]:", p_condic_above_avg)
print(
    "Probabilidad Salario Neto por encima de la Media [Universitario]:", prob_above_avg)
```

**Probabilidad Condicional `salary_monthly_NETO` por Terciarios**

```python
avr_NETO = filtered_df.salary_monthly_NETO.mean()
prob_above_avg = len(
    filtered_df[filtered_df['salary_monthly_NETO'] >= avr_NETO]) / len(filtered_df)

is_above_avg_UNIV = filtered_df['salary_monthly_NETO'] > avr_NETO
univ_bool = filtered_df.profile_studies_level == 'Terciario'

univ_num = filtered_df.loc[filtered_df.profile_studies_level.isin([
                                                                    'Terciario'])]
univ_num = univ_num.profile_studies_level.count()

p_condic_above_avg = len(filtered_df[is_above_avg_UNIV & univ_bool])/univ_num
print(
    "Probabilidad Condicional Salario Neto por encima de la Media [Terciario]:", p_condic_above_avg)
print(
    "Probabilidad Salario Neto por encima de la Media [Terciario]:", prob_above_avg)
```

**Observación:**

Son variables **DEPENDIENTES**
 
Basándonos en el análisis de probabilidad condicional, la probabilidad de ganar por encima de la media siendo Universitario es distinta de la probabilidad de ganar por encima de la media en general. 

Por lo cual, deducimos que son variables **dependientes**.


```python
media_univ = df_Sal_Neto_per_Universitario.salary_monthly_NETO.mean()
media_terc = df_Sal_Neto_per_Terciario.salary_monthly_NETO.mean()
diff_per = (media_univ - media_terc) * 100 / media_univ
print('La diferencia de salarios medios es de:', round(diff_per, 2), '%')
```

**Observación:**

Concluimos que el **salario neto** es **dependiente** del Nivel de Estudio. 

Podemos observar que la media del salario percibido por los Universitarios supera en un **5.39 %** a las tendencias centrales de la subpoblación con el Terciario.
